```python
# 导入HTML
df = pd.DataFrame()
url_list = ['http://www.espn.com/nba/salaries/_/seasontype/4']
for i in range(2, 13):
    url = 'http://www.espn.com/nba/salaries/_/page/%s/seasontype/4'% i
    url_list.append(url)
for url in url_list:
    df = df.append(pd.read_html(url), ignore_index = True)
df = df[[x.startswith('$') for x in df[3]]]
```
