```python
# 
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_excel('data.xls')
# 
x1 = df1['']
y1 = df1['']
y2 = df1['']
y3 = df1['']
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['xtick.direction'] = 'out'     # x
plt.rcParams['ytick.direction'] = 'in'      # y
plt.title('', fontsize = '18')
plt.plot(x1, y1, label = '', color = 'r', marker = 'p')
plt.plot(x1, y2, label = '', color = 'g', marker = '.', mfc = 'r', ms = 8, alpha = 0.7)
plt.plot(x1, y3, label = '', color = 'b', linestyle = '-.', marker = '*')
plt.grid(axis = 'y')                        # y
plt.ylabel('')
plt.yticks(range(50, 150, 10))
plt.legend(['', '', ''])
plt.show()
```
