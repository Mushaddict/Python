```python
# 转换为字典
import pandas as pd
df = pd.read_excel('mrbooks.xls')
df1 = df.groupby(['宝贝标题'])["宝贝总数量"].sum().head()
mydict = df1.to_dict()
print(mydict.items())
for i, j in mydict.items(): 
    print(i, ':/t', j)
```

```python
df.groupby(['宝贝标题'])["宝贝总数量"].describe()
```

```python
# 转换为列表
df1 = df[['宝贝标题']].head()
list1 = df1['宝贝标题'].values.tolist()
for s in list1: 
    print(s)
```

```python
# 转换为元祖
df = pd.read_excel('fl4.xls')
df1 = df[['label1', 'label2']].head()
tuples = [tuple(x) for x in df1.values]
for t in tuples: 
    print(t)
```
